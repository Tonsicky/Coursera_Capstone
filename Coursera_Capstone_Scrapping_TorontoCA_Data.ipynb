```python
#scrapping and parsing data
import requests 
import bs4 
from bs4 import BeautifulSoup
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
canada_postal_data = requests.get(url).text
canada_postal_data_html = BeautifulSoup(canada_postal_data,'html5lib')
print(canada_postal_data_html.prettify())
```

```python
#searching information of interests
tbody_data = canada_postal_data_html.find_all('tbody')
tbody_data_of_interest = tbody_data[0]
td_data_of_interest = tbody_data_of_interest.find_all('td')

borough_list = ['NA','East York','Etobicoke','North York','Toronto','Scarborough','York']
postal_code = []
borough = []
neighbourhood = []

for i in range(len(td_data_of_interest)):
    i_text = td_data_of_interest[i].text
    if "Not assigned" in i_text:
        #postal_code.append(i_text[1:4])
        #borough.append('NA')
        #neighbourhood.append('NA')
        continue 
    else:
        postal_code.append(i_text[1:4])
        #searching for "(" and ")" location
        ob_loc = i_text.rfind('(')
        cb_loc = i_text.rfind(')')
        neighbourhood.append(i_text[ob_loc+1:cb_loc])
        borough.append(i_text[4:ob_loc])
```

```python
#creating dataframe
import pandas as pd 

tnbhood_df = pd.DataFrame({'Postal Code':postal_code,'City/Borough':borough,'Neighbourhood':neighbourhood})

tnbhood_df.head(20)

#tnbhood_df.shape
```
